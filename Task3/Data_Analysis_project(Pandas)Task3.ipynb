```python
import pandas as pd

# Student Data
data = {
    "Name": ["Ali", "Rohan", "John", "Emma", "David", "Eisha", "Rahul", "Priya", "Sophia", "Neha"],
    "Math": [85, 92, 70, 88, 60, 95, 78, 82, 90, 76],
    "Science": [78, 88, 65, 90, 72, 91, 80, 85, 89, 74],
    "English": [90, 95, 75, 85, 68, 93, 79, 87, 92, 80]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save dataset as CSV
df.to_csv("students.csv", index=False)

print("===== STUDENT DATASET =====")
print(df)

# Total Marks
df["Total"] = df["Math"] + df["Science"] + df["English"]

# Average Marks
df["Average"] = df["Total"] / 3

print("\n===== TOTAL & AVERAGE =====")
print(df[["Name", "Total", "Average"]])

# Top Performer
top_student = df.loc[df["Total"].idxmax()]

print("\n===== TOP PERFORMER =====")
print("Name:", top_student["Name"])
print("Total Marks:", top_student["Total"])

# Subject-wise Statistics
print("\n===== SUBJECT STATISTICS =====")

print("Math Average:", df["Math"].mean())
print("Science Average:", df["Science"].mean())
print("English Average:", df["English"].mean())

print("\nHighest Math Marks:", df["Math"].max())
print("Highest Science Marks:", df["Science"].max())
print("Highest English Marks:", df["English"].max())

print("\nLowest Math Marks:", df["Math"].min())
print("Lowest Science Marks:", df["Science"].min())
print("Lowest English Marks:", df["English"].min())

# Save final report
df.to_csv("student_report.csv", index=False)

print("\nAnalysis Completed Successfully!")
print("Files Created:")
print("1. students.csv")
print("2. student_report.csv")
```
